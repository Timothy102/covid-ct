```python
base_model = tf.keras.applications.VGG16()

model = tf.keras.models.Sequential()
model.add(base_model)
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(128, activation = 'relu'))
model.add(tf.keras.layers.Dense(32, activation = 'relu'))

model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Lambda(lambda x: K.sum(x, axis=0,keepdims=True)))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))


model.layers[0].trainable = False

model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
model.build((224,224,3))
model.summary()
```

```python
def loadnib(path):
    x = nib.load(train_path+path)
    return np.array(x.dataobj)
```

```python
def preprocess(path, central_slice = 5):
    image = loadnib(path)
    image = image[:,:, central_slice]
    image =  cv2.resize(image, dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
    img = PIL.Image.fromarray(image).convert('RGB')
    return np.expand_dims(np.asarray(img), axis = 0)
```

```python
def preprocess_batch(path, central_slice = central_slice):
    image = loadnib(path)
    z = image.shape[2]//2
    dz = central_slice//2
    image = image[:,:,z-dz:z+dz]
    a = []
    for k in range(central_slice):
        f =  cv2.resize(image[:,:,k], dsize=(224, 224), interpolation=cv2.INTER_CUBIC)
        f = PIL.Image.fromarray(f).convert('RGB')
        f = np.asarray(f)
        a.append(f)
    return np.asarray(a)
```

```python
ss = []
with open(label_path, 'r') as f:
    for row in f:
        ss.append(row)
```

```python
labels = []
for k in sorted(ss):
    a,b = k.split(',')
    labels.append(int(b))
```

```python
labels = np.array(labels)
labels = np.expand_dims(labels, axis = 1)
labels.shape
```

```python
dataset = tf.data.Dataset.from_generator(generator, output_signature=(
         tf.TensorSpec(shape=(6, 224,224,3), dtype=tf.float32),
         tf.TensorSpec(shape=(1), dtype=tf.int32)))

dataset.repeat(10).batch(8).shuffle(10000)
```

```python
from tensorflow.keras import Sequential, Model, Input
from tensorflow.keras.layers import * 
```

```python
INPUT_SHAPE = (224, 224, 3)
```

```python
cnn_model = Sequential([
    Conv2D(64, (3,3), activation = 'relu'),
    MaxPooling2D((2,2)),
    Conv2D(32, (3,3), activation = 'relu'),
    MaxPooling2D((2,2)),
    Flatten(),
    Dense(16, activation = 'relu'),
    Dropout(0.1),
    Lambda(lambda x: K.sum(x, axis=0,keepdims=True)),

    Dense(1, activation = 'sigmoid')
])

cnn_model.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
cnn_model.build((None, 224,224,3))
cnn_model.summary()
```

```python
filepath = 'vgg_model/'
test_dataset = dataset.take(300) 
train_dataset = dataset.skip(55)
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, ModelCheckpoint
reduce_lr = ReduceLROnPlateau(monitor = 'loss')
check = ModelCheckpoint(filepath, save_weights_only = True, save_best_only = True)
history = cnn_model.fit(train_dataset, validation_data = test_dataset,  epochs = 5, verbose = 1, callbacks = [reduce_lr, check])
```
